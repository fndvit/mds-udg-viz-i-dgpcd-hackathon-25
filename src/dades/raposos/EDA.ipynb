# Tractament de les dades

Observem que hi ha columnes de valor únic (`solar_termica`,`solar_fotovoltaica`).

Observem també anomalies en la columna `us_edifici`, amb tota una serie de valors equivalents comptats com a diferents per canvis ideomàtics o de picatge. N'arreglem els valors.

L'`any_construccio` es llegeix com a `float`, caldria converir-ho a `int`. Però com que hi ha `nan`s, no admet la transformació. Fem un `fillna` per assignar-li un valor i canviem el tipus de la columna.

```python
# df.drop(columns=['solar_termica', 'solar_fotovoltaica'], inplace=True)
df.loc[(df['us_edifici']=='Vivienda unifamiliar') | (df['us_edifici']=='Habitatge Unifamiliar'), "us_edifici"] = "Habitatge unifamiliar"
df.loc[df['us_edifici']=='Vivienda individual en bloque de viviendas', "us_edifici"] = "Habitatge individual en bloc d'habitatges"
df.loc[df['us_edifici']=='Terciario',"us_edifici"] = 'Terciari'
df.loc[df['us_edifici']=='Bloque de viviendas',"us_edifici"] = "Bloc d'habitatges"
df.loc[df['us_edifici']=='Bloque de viviendas plurifamiliar' ,"us_edifici"] = "Bloc d'habitatges plurifamiliar"
df['any_construccio'] = df['any_construccio'].fillna(0).astype(int)
```

Codifiquem les variables binàries (0: NO, 1: SI)

```python
df.loc[df['zona_climatica'] == "SI",'zona_climatica'] = 1
df.loc[df['zona_climatica'] == "NO",'zona_climatica'] = 0
df.loc[df['any_construccio'] == "SI",'any_construccio'] = 1
df.loc[df['any_construccio'] == "NO",'any_construccio'] = 0
df.loc[df['us_edifici'] == "SI",'us_edifici'] = 1
df.loc[df['us_edifici'] == "NO",'us_edifici'] = 0
df.loc[df['vehicle_electric'] == "SI",'vehicle_electric'] = 1
df.loc[df['vehicle_electric'] == "NO",'vehicle_electric'] = 0
df.loc[df['solar_termica'] == "SI",'solar_termica'] = 1
df.loc[df['solar_termica'] == "NO",'solar_termica'] = 0
df.loc[df['solar_fotovoltaica'] == "SI",'solar_fotovoltaica'] = 1
df.loc[df['solar_fotovoltaica'] == "NO",'solar_fotovoltaica'] = 0
df.loc[df['sistema_biomassa'] == "SI",'sistema_biomassa'] = 1
df.loc[df['sistema_biomassa'] == "NO",'sistema_biomassa'] = 0
df.loc[df['xarxa_districte'] == "SI",'xarxa_districte'] = 1
df.loc[df['xarxa_districte'] == "NO",'xarxa_districte'] = 0
df.loc[df['energia_geotermica'] == "SI",'energia_geotermica'] = 1
df.loc[df['energia_geotermica'] == "NO",'energia_geotermica'] = 0
df.loc[df['rehabilitacio_energetica'] == "SI",'rehabilitacio_energetica'] = 1
df.loc[df['rehabilitacio_energetica'] == "NO",'rehabilitacio_energetica'] = 0
```

```python
print(f"zona_climatica: {df['zona_climatica'].unique()}")
print(f"any_construccio: {df['any_construccio'].unique()}")
print(f"us_edifici: {df['us_edifici'].unique()}")
print(f"vehicle_electric: {df['vehicle_electric'].unique()}")
print(f"solar_termica: {df['solar_termica'].unique()}")
print(f"solar_fotovoltaica: {df['solar_fotovoltaica'].unique()}")
print(f"sistema_biomassa: {df['sistema_biomassa'].unique()}")
print(f"xarxa_districte: {df['xarxa_districte'].unique()}")
print(f"energia_geotermica: {df['energia_geotermica'].unique()}")
print(f"rehabilitacio_energetica: {df['rehabilitacio_energetica'].unique()}")
```

Hi ha un error en la generació dels codis de la columna `mundissec`. Per com es generen, sobra un dels caràcters i, com a resultat de lectura del dataset, s'assigna el tipus float. Per tant, fem el següent:
* Transformem els nombres a string, assegurant-nos que tenen el format correcte.
* Eliminem el dígit sobrant de cada codi no null.

```python
df['mundissec'] = df['mundissec'].apply(lambda x: ("%.0f" % x).zfill(11) if not pd.isna(x) else x)
df['mundissec'] = df['mundissec'].apply(lambda x: x[:5] + x[6:] if not pd.isna(x) else x)
```
